```python
import pandas as pd
import seaborn as sns
```

```python
df_writes = pd.read_csv('Data/writes.csv', sep=';')

df_writes['LastWriteTime'] = pd.to_datetime(df_writes['LastWriteTime'], format='%d.%m.%Y %H:%M:%S')

df_writes
```

```python
# Create a heatmap of the write occurrences
sns.heatmap(df_writes.groupby([df_writes['LastWriteTime'].dt.date, df_writes['LastWriteTime'].dt.hour]).size().unstack(), cmap='viridis')
```

```python
# Create a KDE plot of the write occurrences
df_writes_after_2020 = df_writes[df_writes['LastWriteTime'].dt.year >= 2020]

sns.kdeplot(df_writes_after_2020['LastWriteTime'], fill=True, color='red')
```

```python
# Create a KDE plot of the write occurrences in an average day after 2020

sns.kdeplot(df_writes_after_2020['LastWriteTime'].dt.hour, fill=True, color='red')
```
