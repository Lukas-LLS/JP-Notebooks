```python
import pandas as pd
import seaborn as sns
```

# Dew Points

This dataset contains the dew points for specific temperatures and humidity levels.
The data was calculated using magnus formula.

```python
df_dew_points = pd.read_csv("Data/dewPoints.csv")
df_dew_points
```

# Visualizations

The following visualizations show the distribution of the dew points for the temperature range 0°C to 50°C.

```python
sns.histplot(df_dew_points, x="Dew Point", bins=20, color="lightblue", kde=True)
```

```python
# Show all humidity levels and temperatures that result in a dew point of 20°C (±0.5°C) in a plot
sns.scatterplot(data=df_dew_points[(df_dew_points["Dew Point"] >= 19.5) & (df_dew_points["Dew Point"] <= 20.5)], x="Temperature", y="Relative Humidity", color="lightblue")
```

```python
# Create a heatmap of the dew points for the temperature range 0°C to 50°C
df_dew_points_heatmap = df_dew_points.pivot(columns=["Temperature"], index=["Relative Humidity"], values="Dew Point")
sns.heatmap(df_dew_points_heatmap, cmap="coolwarm")
```
